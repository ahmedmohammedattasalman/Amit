```python

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode(0)
    cur = dummy
    carry = 0
    p, q = l1, l2
    while p or q or carry:
        x = p.val if p else 0
        y = q.val if q else 0
        s = x + y + carry
        carry = s // 10
        cur.next = ListNode(s % 10)
        cur = cur.next
        if p: p = p.next
        if q: q = q.next
    return dummy.next

def to_listnode(lst):
    dummy = ListNode(0)
    cur = dummy
    for v in lst:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

def to_list(node):
    res = []
    while node:
        res.append(node.val)
        node = node.next
    return res

l1 = to_listnode([2,4,3])
l2 = to_listnode([5,6,4])
print(to_list(addTwoNumbers(l1, l2))) 


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode(0)
    cur = dummy
    carry = 0
    p, q = l1, l2
    while p or q or carry:
        x = p.val if p else 0
        y = q.val if q else 0
        s = x + y + carry
        carry = s // 10
        cur.next = ListNode(s % 10)
        cur = cur.next
        if p: p = p.next
        if q: q = q.next
    return dummy.next

def to_listnode(lst):
    dummy = ListNode(0)
    cur = dummy
    for v in lst:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

def to_list(node):
    res = []
    while node:
        res.append(node.val)
        node = node.next
    return res


l1 = to_listnode([2,4,3])
l2 = to_listnode([5,6,4])
print(to_list(addTwoNumbers(l1, l2)))  
```
