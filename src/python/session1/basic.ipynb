```python
# task_1
email = "Amit_ml@gmail.edu"


'''# 1 - Input Validation: Check if the input string contains exactly one "@" symbol and
at least one "." after the "@" symbol. If it's not a valid email, return "Invalid
email".
'''
checker=0

for check in email:
    if ("@" in check):
      checker+=1
if(checker>1 and "." not in email):
   print("invalid email")
elif (checker==1 and "." in email):
   print("valid email")
   
   
#2 - Extract Username: Extract and return the part of the email before the "@"
#symbol.

at_checker = email.find("@")
username= email[0:at_checker]
print(username)

'''3- Extract Domain: Extract and return the domain (the part between "@" and the
last ".").'''
 
print(email[email.find("@")+1:email.find(".")])

if email[email.find("."):] == ".edu":
   print("educational")
elif email[email.find("."):] == ".com":
   print("commercial ")
else:
   print("other domain")

```

```python
massage = "&&&**$gnirtS PLIO!!@1234"


'''
1. Extract the core part of the message: "gnirtS PLIO".
2. Reverse the first word: "gnirtS" becomes "String".
3. Replace shifted vowels in the second word:
o "PLIO": Replace I->E and O->U to get "PLEU".
4. Final decoded message: "String PLEU".
'''

core_string = massage[massage.find('$')+1 : massage.find("!")]
first= core_string[0:core_string.find(" ")][::-1]
second = core_string[core_string.find(" ")+1:17]

second=second.replace('I' , 'E')
second=second.replace('O' , 'U')

print( first,second)
```
