```python
def fac(num):


    coun =1
    while(num!=0):
        coun*=num
        num-=1
    return coun



print(fac(3))
```

```python
def div (mn , mx):
    min_1=min(mn,mx)
    list_1 =[]
    for i in range(1, min_1+1):
        if mn % i == 0 and mx % i ==0:
            list_1.append(i)
```
