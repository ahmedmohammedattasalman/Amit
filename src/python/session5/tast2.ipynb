```python
import os
import random

def thanos_project(thanos, num_files):
  
    if not os.path.exists(thanos):
        os.mkdir(thanos)

   
    for i in range(num_files):
        file_path = os.path.join(thanos, f"file_{i}.txt")
        with open(file_path, "w") as f:
            f.write("new line \n")
    
    files = os.listdir(thanos)
   
    half = len(files) // 2
    files_to_delete = random.sample(files, half)

    for file in files_to_delete:
        os.remove(os.path.join(thanos, file))

    # 5- check result
    final_files = os.listdir(thanos)
    print("result is :", len(final_files), "files")


thanos_project("thanos_folder", 10)

```
