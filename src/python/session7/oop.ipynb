```python
class Calculator1:
    def summation(self,n1,n2):
        return n1+n2
    def subtraction(self, n1 , n2):
        return n1-n2
    def multiplication(self, n1,n2):
        return n1*n2
    def division(self,n1,n2):
        return n1/n2
class Calculator2(Calculator1):
    def power(self, n1,n2):
        return n1**n2
c1= Calculator2()
print(c1.division(6,3))   
```

```python
from prettytable import  PrettyTable
products = [
    {"name":"water","price":"80","quantity":"1500"},
    {"name":"soda","price":"130","quantity":"1500"},
    {"name":"chips","price":"75","quantity":"1500"},
    {"name":"bread","price":"45","quantity":"1500"},
    {"name":"eggs","price":"200","quantity":"1500"}
            
            ]
table = PrettyTable()
table.field_names=["products name" , "price($)","qunatity" ]
for p in products:
    table.add_row([p["name"],f"{p['price']}",p['quantity']])


print(table)

total_bill = 0
while True:
    product_name = input("\n enter product name or E to finsh").strip().lower()
    if product_name =='E':
        break
    product = next((p for p in products if p['name'].lower() == product_name) , None)
    if product is None:
        print("product not found! try again ")
        continue
    try:
        quantity_rquired = float(input(f"how many {product['name']} do you want ?"))
        if quantity_rquired<=0:
            print("quantity must be more than zero")
            continue
        if quantity_rquired > products["quantity"]:
            print(f"insufficinet quantity !")
            continue
        dicount_persentage = 0.05
        price_before_disc  = quantity_rquired * products['price']
        dicount_amount = price_before_disc - dicount_persentage
        final_price = price_before_disc - dicount_amount
        products['quantity'] -= quantity_rquired
        total_bill += final_price
        print(f"success! price fpr {products['name']}: ${final_price:.2f} (saved: ${dicount_amount:.2f})")
        another = input("something else? (y/n)").lower()
        if another != 'y' :
            break
    except ValueError:
        print("invalid input !")
        print(f"\n total amount to pay : $ { total_bill:.2f}")
    
    



```
